```python
#depending on the states of the top 5 country
query_5 = """WITH Country_Revenue AS (
    SELECT 
        Country,
        SUM(Quantity * Unit_Price_USD) AS Total_Revenue
    FROM 
        alltable
    GROUP BY 
        Country
    ORDER BY 
        Total_Revenue DESC
    LIMIT 5
),
State_Revenue AS (
    SELECT 
        Country,
        State,
        SUM(Quantity * Unit_Price_USD) AS State_Revenue
    FROM 
        alltable
    WHERE 
        Country IN (SELECT Country FROM Country_Revenue)  -- Filter for top 5 countries
    GROUP BY 
        Country, State
)
SELECT 
    Country,
    State,
    State_Revenue
FROM 
    State_Revenue
ORDER BY 
    State_Revenue DESC;"""
```

```python
mycursor.execute(query_5)
results = mycursor.fetchall()
for row in results:
    print(row)
```

```python
#age grouping
query_6 = """SELECT 
    CASE
        WHEN FLOOR(DATEDIFF(CURDATE(), Birthday) / 365) BETWEEN 0 AND 18 THEN '0-18'
        WHEN FLOOR(DATEDIFF(CURDATE(), Birthday) / 365) BETWEEN 19 AND 30 THEN '19-30'
        WHEN FLOOR(DATEDIFF(CURDATE(), Birthday) / 365) BETWEEN 31 AND 45 THEN '31-45'
        WHEN FLOOR(DATEDIFF(CURDATE(), Birthday) / 365) BETWEEN 46 AND 60 THEN '46-60'
        ELSE '60+'
    END AS Age_Category,
    COUNT(*) AS People_Count,  -- Count the number of people in each age group
    SUM(Quantity * Unit_Price_USD) AS Total_Revenue  -- Calculate the total revenue for each age group
FROM 
    alltable
GROUP BY 
    Age_Category
ORDER BY 
    Total_Revenue DESC;  -- Order by total revenue in descending order
"""
```

```python
mycursor.execute(query_6)
results = mycursor.fetchall()
for row in results:
    print(row)
```

```python
#from which agegroup
query_12 = """SELECT 
    Age_Group,
    SUM(Revenue) AS Total_Revenue
FROM (
    SELECT 
        FLOOR(DATEDIFF(CURRENT_DATE, Birthday) / 365) AS Age, -- Calculate age
        Quantity * Unit_Price_USD AS Revenue,  -- Calculate revenue for each row
        CASE 
            WHEN FLOOR(DATEDIFF(CURRENT_DATE, Birthday) / 365) >= 0 
                AND FLOOR(DATEDIFF(CURRENT_DATE, Birthday) / 365) < 15 THEN '0-15'
            WHEN FLOOR(DATEDIFF(CURRENT_DATE, Birthday) / 365) >= 15 
                AND FLOOR(DATEDIFF(CURRENT_DATE, Birthday) / 365) < 30 THEN '15-30'
            WHEN FLOOR(DATEDIFF(CURRENT_DATE, Birthday) / 365) >= 30 
                AND FLOOR(DATEDIFF(CURRENT_DATE, Birthday) / 365) < 45 THEN '30-45'
            WHEN FLOOR(DATEDIFF(CURRENT_DATE, Birthday) / 365) >= 45 
                AND FLOOR(DATEDIFF(CURRENT_DATE, Birthday) / 365) < 60 THEN '45-60'
            ELSE '60+'  -- Group those who are 60+ years old
        END AS Age_Group
    FROM alltable
) AS AgeData  -- Subquery that calculates age and revenue, and assigns an age group
GROUP BY Age_Group  -- Group by the calculated age group
ORDER BY Total_Revenue DESC;  -- Order by the total revenue in descending order
"""
```

```python
mycursor.execute(query_12)
results = mycursor.fetchall()
for row in results:
    print(row)
```
